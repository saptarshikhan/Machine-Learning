```python
df=pd.read_csv(r'D:/dataset/company_sales_data.csv')
df.set_index('month_number')
```

```python
#Read Total profit of all months and show it using a line plot
plt.bar(df['month_number'],df['total_profit'],width=0.5,color='red')
plt.xlabel("Month_number")
plt.ylabel("total_profit")
plt.title("total profit in every month")
plt.show()
```

```python
#Get total profit of all months and show line plot with the following Style properties
plt.plot(df['month_number'],df['total_profit'],marker='o',color='red',linewidth=3.0)
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(['sales/month'],loc='lower right')
plt.show()
```

```python
#Read all product sales data and show it using a multiline plot
plt.bar(df['month_number'],df['facecream'],width=0.1)
plt.bar(df['month_number']+0.1,df['facewash'],width=0.1)
plt.bar(df['month_number']+0.2,df['toothpaste'],width=0.1)
plt.bar(df['month_number']+0.3,df['bathingsoap'],width=0.1)
plt.bar(df['month_number']+0.4,df['shampoo'],width=0.1)
plt.bar(df['month_number']+0.5,df['moisturizer'],width=0.1)
plt.legend(labels=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'])
plt.show()
```

```python
#: Read toothpaste sales data of each month and show it using a scatter plot
plt.scatter(df['month_number'],df['toothpaste'])
plt.grid(linestyle='-')
```

```python
#Read face cream and facewash product sales data and show it using the bar chart
plt.bar(df['month_number'],df['facecream'],width=0.25)
plt.bar(df['month_number']+0.25,df['facewash'],width=0.25)
plt.legend(labels=['facecream','facewash'])
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
```

```python
# Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk
plt.bar(df['month_number'],df['bathingsoap'],color='pink')
plt.legend(labels=['bathingsoap'])
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.savefig('bathingsoap.png')
```

```python
#Read the total profit of each month and show it using the histogram to see the most common profit ranges
plt.hist(df['total_profit'],bins=10)
```

```python
#Calculate total sale data for last year for each product and show it using a Pie chart
plt.pie([df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()],labels=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'])
```

```python
#Read Bathing soap facewash of all months and display it using the Subplot

plt.subplot(121)
plt.plot(df['month_number'],df['bathingsoap'])

plt.subplot(122)
plt.plot(df['month_number'],df['facewash'])

plt.show()
```

```python
#Read all product sales data and show it using the stack plot
plt.stackplot(df['month_number'], df['facecream'],df['facewash'],df['toothpaste'],df['bathingsoap'],df['shampoo'],df['moisturizer'])
plt.xlabel('No of months')
```
