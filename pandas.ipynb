```python
#show 10 prime numbers in series format
def prime (n):
    c=0
    for i in range(1,n):
        if n%i==0:
            c+=1
    if c==1:
        return True
    return False
l=[]
i=1
n=1
while i<=10:
    if prime(n)==True:
        i+=1
        l.append(n)
        n+=1
    else :
        n+=1
print(l)
s=pd.Series(l,index=[i for i in range(1,11)])
print(s)
```

```python
url='https://en.wikipedia.org/wiki/Main_Page'
resp=requests.get(url)
s=resp.text.lower()
#soup=BeautifulSoup(resp.text,'html.parser')
#print(soup)
d={'a':0,'e':0,'i':0,'o':0,'u':0}
for i in s:
    if i in d:
        d[i]+=1
x=pd.Series(d)
print(x)
```

```python
plt.pie(d.values(),labels=d.keys(),autopct='%.2f%%',colors=['red','green','blue','yellow','pink'],explode=[0.1,0,0,0,0])
plt.show()
```

```python
df=pd.read_csv(r'D:/Dataset/reasult.csv',index_col=0)
df.head()
```

```python

#finding total of all the students
df['total']=df['Eng']+df['Maths']+df['Sci']

#finding avarage of all the students
df['persent']=df['total']//3
df.head()
```

```python

#finding grades of all the students
x=df['persent']
a=[]
for i in x:
    if i>=90:
        g='A+'
    elif i>=80:
        g='A'
    elif i>=70:
        g='B'
    elif i>=60:
        g='C'
    elif i>=50:
        g='D'
    elif i>=35:
        g='P'
    else:
        g='F'
    a.append(g)
df['Grade']=a
df
```
